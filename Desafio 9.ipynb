### Desafio 9

Escreva uma função que retorne a soma dos múltiplos de 3 e 5 entre 0 e um número limite (parâmetro). Por exemplo, se o limite for 20, ele retornará a soma de 3, 5, 6, 9, 10, 12, 15, 18, 20.

```python
def somas(entry):
    counter = 0
    for x in range(entry+1):
        if x%3 == 0:
            counter+=x
        elif x%5 == 0:
            counter+=x
        else:
            pass
    return counter

preta = somas(5)
print(preta)
```
