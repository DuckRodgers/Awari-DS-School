### Desafio 10

Dada uma lista, divida-a em 3 partes iguais e reverta cada lista.

Exemplo: 

sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

Parte 1 [8, 45, 11]

Parte 2 [12, 14, 23]

Parte 3 [89, 45, 78]

```python
sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

def split_list(entry):
    longs = len(entry)
    longs /= 3
    longs = int(longs)
    for x in range(longs):
        list1 = entry[:longs]
        list2 = entry[longs:longs*2]
        list3 = entry[longs*2:]
    list1 = list1[::-1]
    list2 = list2[::-1]
    list3 = list3[::-1]
    return list1, list2, list3

x,y,z = split_list(sampleList)
```
