Escreva um script Python para encontrar as 10 palavras mais longas em um arquivo.

O arquivo TXT está localizado na mesma pasta do projeto (texto.txt).

```python
#The code isn't perfect. There are two obvious mistakes: "language.its" is filtered because he considers ".". "general-purpose" happens for the same reason. On the other hand, "intermediate" I have no clue why it happens.

edit = "/text.txt"
org = edit

def my_len(n):
    return len(n)

with open(org,'r') as file:
    res = []
    for lines in file:
        split_list = lines.split(" ")
        for split in split_list:
            if split == '' or split == '\n':
                pass
            else:
                res.append(split)

sol = sorted(res,key=lambda x: len(x), reverse=True)
sol = sol[:10]
print(sol)
```
